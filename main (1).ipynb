## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
    """Calculate the factorial of a non-negative integer n."""
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop,
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n.
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function
test_factorial(factorial)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ğŸ§`

```python
def mode_counter(arr):
#your code here
  frequency = {}
  for item in arr:
          if item in frequency:
              frequency[item] += 1
          else:
              frequency[item] = 1
  max_count = -1
  mode = None
  for item, count in frequency.items():
          if count > max_count:
              max_count = count
              mode = item

  return mode
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ğŸ˜‰`

```python
def st_dev(list_sd):
#your code here
    n = len(list_sd)
    # Calculate the mean
    mean = sum(list_sd) / n

    # Calculate the sum of squared differences from the mean
    sum_squared_diffs = sum((x - mean) ** 2 for x in list_sd)

    # Calculate the variance using n-1
    variance = sum_squared_diffs / (n - 1)

    # Calculate the square root of variance
    return variance ** 0.5
```

```python
# This will test your function
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
#your code here
    # Create a set of all alphabets
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    # Normalize the string to lowercase and filter out non-alphabet characters
    processed_string = set(char.lower() for char in string if char.isalpha())
    # Check if the processed string contains all the alphabets
    return alphabet <= processed_string
```

```python
# This will test your function
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ğŸ¤ª`

```python
def sort_alpha(string):
#your code here
  words = [word.strip() for word in string.split(',')]

  # Sort the list of words
  sorted_words = sorted(words)

  # Using string concatenation in a loop to emulate the join operation
  sorted_string = ",".join(sorted_words)

  return sorted_string
```

```python
# This will test your function
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here
    if len(password) < 8:
        return False

    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif not char.isalnum():
            has_special = True

    # Check if all conditions are met to stop early for efficiency
    if has_lower and has_upper and has_digit and has_special:
          return True

    return has_lower and has_upper and has_digit and has_special
```

```python
# This will test your function
test_pass(check_pass)
```
