```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generate_uniform(bottom, ceiling, count):
    """
    Generates uniformly distributed random numbers within a specified range.
    
    Parameters:
    - bottom: The lower boundary of the range.
    - ceiling: The upper boundary of the range.
    - count: The number of random numbers to generate.
    
    Returns:
    - An array of uniformly distributed random numbers.
    """
    return uniform(loc=bottom, scale=ceiling - bottom).rvs(count)

# Generate random numbers for the given parameters
data1 = generate_uniform(bottom=10, ceiling=15, count=100)
data2 = generate_uniform(bottom=10, ceiling=60, count=1000)

# Create histograms
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot for the first set of parameters
ax[0].hist(data1, bins=10, color='blue', edgecolor='black')
ax[0].set_title('Uniform Distribution: bottom=10, ceiling=15, count=100')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Count')

# Plot for the second set of parameters
ax[1].hist(data2, bins=10, color='green', edgecolor='black')
ax[1].set_title('Uniform Distribution: bottom=10, ceiling=60, count=1000')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
#The first distribution ranges from 10 to 15 with 100 samples, resulting in a narrow, slightly varied histogram. The second spans from 10 to 60 with 1,000 samples, showing a broader, more uniform distribution with each bin covering a larger value range, ensuring an even spread of data across bins.
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

```python
from scipy.stats import norm 
```

```python
def generate_normal(mean, std, count):
    """
    Generates normally distributed random numbers using scipy.stats.norm.
    
    Parameters:
    - mean: The mean (average) of the distribution.
    - std: The standard deviation of the distribution.
    - count: The number of random numbers to generate.
    
    Returns:
    - An array of normally distributed random numbers.
    """
    return norm.rvs(loc=mean, scale=std, size=count)

# Generate normal distributions
data1 = generate_normal(mean=10, std=1, count=1000)
data2 = generate_normal(mean=10, std=50, count=1000)

# Create histograms
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot for the first distribution
ax[0].hist(data1, bins=30, color='blue', edgecolor='black')
ax[0].set_title('Normal Distribution: Mean=10, SD=1')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Count')

# Plot for the second distribution
ax[1].hist(data2, bins=30, color='green', edgecolor='black')
ax[1].set_title('Normal Distribution: Mean=10, SD=50')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_exponential(mean, size):
    """
    Generates exponentially distributed random numbers with a specified mean.
    
    Parameters:
    - mean: The mean (average) of the distribution.
    - size: The number of random numbers to generate.
    
    Returns:
    - An array of exponentially distributed random numbers.
    """
    # The scale parameter for an exponential distribution is the mean itself.
    scale = mean
    return np.random.exponential(scale=scale, size=size)

# Generate two sequences of exponentially distributed numbers
data1 = generate_exponential(mean=10, size=10)
data2 = generate_exponential(mean=10, size=100)

# Create histograms
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Histogram for the first sequence
ax[0].hist(data1, bins=100, color='blue', edgecolor='black')
ax[0].set_title('Exponential Distribution: Size=10')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Frequency')

# Histogram for the second sequence
ax[1].hist(data2, bins=100, color='green', edgecolor='black')
ax[1].set_title('Exponential Distribution: Size=100')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Frequency')

plt.tight_layout()  # Adjust the layout to make room for all the subplots
plt.show()  # Display the plots
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
from scipy.stats import expon

# Set the scale parameter (mean) of the exponential distribution
scale = 10  # mean = 10 minutes

# Calculate the probability of spending less than 15 minutes in the bank
probability = expon.cdf(15, scale=scale)
print(f"The probability that a customer will spend less than 15 minutes in the bank is {probability:.4f}")
```
