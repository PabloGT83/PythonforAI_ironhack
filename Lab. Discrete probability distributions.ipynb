## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
# Number of apples and oranges
num_apples = 60
num_oranges = 40
total_fruits = num_apples + num_oranges  # Total number of fruits

# Probability of picking an apple
p_apple = num_apples / total_fruits
# Probability of picking an orange
p_orange = num_oranges / total_fruits

print(f"The probability that the fruit is an apple is {p_apple:.2f}")
print(f"The probability that it is an orange is {p_orange:.2f}")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# Probabilities of picking an apple and an orange
p = 0.6
q = 0.4

# Probability of the first 5 fruits being all apples
prob_5_apples = p**5

# Probability of the first 5 fruits being all apples and the next 15 being all oranges
prob_5_apples_15_oranges = (p**5) * (q**15)

print(f"The probability that the first 5 fruits are all apples is {prob_5_apples:.5f}")
print(f"The probability that the first 5 fruits are all apples and the next 15 fruits are all oranges is {prob_5_apples_15_oranges:.10f}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
from scipy.stats import binom

# Number of trials and number of successes
n = 20
k = 5
# Probability of success (picking an apple)
p = 0.6

# Calculate the probability of exactly 5 apples in 20 picks
prob_5_apples = binom.pmf(k, n, p)

print(f"The probability of getting exactly 5 apples and 15 oranges in 20 picks is {prob_5_apples:.5f}")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
from scipy.stats import binom

# Number of trials
n = 20
# Probability of picking an apple
p = 0.6

# Calculate the probability of getting fewer than 5 apples
prob_less_than_5_apples = binom.cdf(4, n, p)  # Using 4 because we need P(X < 5), which is cdf at 4

print(f"The probability of picking fewer than 5 apples in 20 tries is {prob_less_than_5_apples:.5f}")
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 20  # total number of trials
p = 0.6  # probability of success on each trial (picking an apple)

# Generating the range of possible outcomes from 0 to n
x = np.arange(0, n+1)  # x values from 0 to n (inclusive)

# Calculating the probability mass function for each outcome
pmf_values = binom.pmf(x, n, p)

# Plotting the probability distribution function (PDF)
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.title('Probability Distribution Function (PDF) for Binomial Distribution')
plt.xlabel('Number of Apples Picked')
plt.ylabel('Probability')
plt.xticks(x)  # Ensures each integer value is marked
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Adding grid lines for better readability

plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math

# Parameters for the Poisson distribution
lambda_ = 2.3  # Average number of goals per match
k = 5  # Number of goals for which we are calculating the probability

# Calculate the Poisson probability using the formula
probability = (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

print(f"The probability that the number of goals scored in a match is 5 is {probability:.5f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
import matplotlib.pyplot as plt
import math

# Average number of goals per match (Poisson lambda parameter)
lambda_ = 2.3

# Range of number of goals from 0 to 10
x = range(0, 11)  # Possible number of goals

# Compute Poisson probabilities for each k in x
probabilities = [(lambda_**k * math.exp(-lambda_) / math.factorial(k)) for k in x]

# Plotting the Poisson distribution
plt.bar(x, probabilities, color='blue', alpha=0.7)
plt.title('Poisson Probability Distribution of Goals per Match')
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.xticks(x)  # Ensures each integer value is marked
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Adding grid lines for better readability

plt.show()
````
